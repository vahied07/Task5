```python
# 📦 Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 🎨 Style and Display Settings
sns.set(style="whitegrid")
%matplotlib inline

# 📁 Load Dataset
df = pd.read_csv(r"C:\Users\jagad\OneDrive\Documents\tested.csv")

# 🧾 Basic Info
print("🔹 Dataset Info")
df.info()

print("\n🔹 Descriptive Statistics")
print(df.describe(include="all"))

print("\n🔹 Missing Values")
print(df.isnull().sum())

# 🔧 Handle Missing Values
df['Age'].fillna(df['Age'].median(), inplace=True)
df.dropna(subset=['Embarked'], inplace=True)

# 📊 Univariate Analysis

# Gender Distribution
print("\n🔹 Gender Counts")
print(df['Sex'].value_counts())
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()

# Class Distribution
print("\n🔹 Class Counts")
print(df['Pclass'].value_counts())
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()

# Survival Distribution
print("\n🔹 Survival Counts")
print(df['Survived'].value_counts())
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()

# Embarked Distribution
print("\n🔹 Embarkation Counts")
print(df['Embarked'].value_counts())
sns.countplot(x='Embarked', data=df)
plt.title('Embarkation Distribution')
plt.show()

# Age Distribution (Histogram)
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Age Boxplot
sns.boxplot(x=df['Age'])
plt.title('Age Boxplot')
plt.show()

# 📈 Bivariate Analysis

# Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()

# Survival by Class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival by Class')
plt.show()

# 🔗 Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 🔄 Pairplot of Key Features
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle('Pairplot: Age, Fare, Pclass vs Survival', y=1.02)
plt.show()
```
